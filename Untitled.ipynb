```python
l40 = f.layers[40]
```

```python
v = []; c = []
lines = iter(f.layers[40].lines)
l = next(lines)

#Find first location and move to it
while not(l.code in ('G0', 'G1') and ('X' in l.args or 'Y' in l.args)):
    l = next(lines)
v.append((l.args['X'], l.args['Y']))
c.append(Path.MOVETO)

for l in lines:
    if l.code in ('G0', 'G1') and ('X' in l.args or 'Y' in l.args):
        v.append((l.args['X'], l.args['Y']))
        if 'E' in l.args and l.args['E'] > 0:
            c.append(Path.LINETO)
        else:
            c.append(Path.MOVETO)
```

```python
path = Path(v,c)
fig, ax = subplots()
patch = patches.PathPatch(path, fc='none', lw=2, ec='b')
ax.add_patch(patch)
xlim(84, 95)
ylim(84, 95)
ax.set_aspect('equal')
```
